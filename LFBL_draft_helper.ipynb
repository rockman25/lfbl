```python
def update(player='XYZ', owner='Beans'):
    stats.loc[player,'OWNER']=owner

def add_rookie(player='XYZ'):
    stats.loc[player,:-1]=np.nan
    stats.loc[player,'OWNER']='NONE'

def sort(column='PTS'):
    frame = stats.sort_values(by=column, ascending=False)
    return frame[frame['OWNER']=='NONE'][[column,'NBA_FANTASY_PTS_RANK','OWNER']]

def show_stats(players=[]):
    frame = stats.sort_values(by='NBA_FANTASY_PTS', ascending=False)
    frame = frame[frame['OWNER']=='NONE']
    if len(players)>0:
        return frame.loc[players,['NBA_FANTASY_PTS_RANK','PTS','FG3M','REB','AST','BLK','STL','TOV','OWNER']]
    else:
        return frame.loc[:,['NBA_FANTASY_PTS','PTS','FG3M','REB','AST','BLK','STL','TOV','OWNER']]
```
